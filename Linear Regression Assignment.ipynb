## 3: Data Preparation

```python
# Creating Dummy variable for months, weekdays, weathersit and season variables.

Months_dt = pd.get_dummies(bike_data.month,drop_first=True)
Weekdays_dt = pd.get_dummies(bike_data.weekday,drop_first=True)
Weathersit_dt = pd.get_dummies(bike_data.weathersit,drop_first=True)
Season_dt = pd.get_dummies(bike_data.season,drop_first=True)
```

# 4: Splitting the Data into Training and Testing Sets.

```python
# Splitting the dataframe into Train and Test

np.random.seed(0)
Dt_train, Dt_test = train_test_split(bike_data_new, train_size = 0.7, random_state = 100)
```

```python
# Checking the shape of training Dataframe.
Dt_train.shape
```

```python
# Checking the shape of testing Dataframe.
Dt_test.shape
```

```python
# Using MinMaxScalar to Rescaling the features.
scaler = MinMaxScaler()
```

```python
# Verifying the head of Dataframe before Scaling.

Dt_train.head()
```

```python
# Applying scaler() to all the columns except the 'yes-no', and 'dummy' variables.
var_num = ['temp','atemp','humidity','windspeed','cnt']
Dt_train[var_num] = scaler.fit_transform(Dt_train[var_num])
```

```python
# Now Verifying the head after applying the scaling.

Dt_train.head()
```

```python
# Describing the Data_set.
Dt_train.describe()
```

```python
 # Checking the correlation coefficients to see which variables are highly correlated.
    
plt.figure(figsize = (30,30))
matrix = np.triu(Dt_train.corr())
sns.heatmap(Dt_train.corr(), annot = True, cmap="RdYlGn", mask=matrix )
plt.show()
```

##### cnt seems to have year variable and temp. Similarly, humidity and Misty shows correlation. Spring season with Jan and Feb and Summer season with May and Winter season with Oct and Nov results a good correlation.

```python
# Visualizing one of the correlaton to see the trends using Scatter plot.

plt.figure(figsize=[7,6])
plt.scatter(Dt_train.temp, Dt_train.cnt)
plt.title("Correlation Between temp And cnt", fontsize = 12)
plt.show()
```

##### The above plot confirms a  positive correlation between temp and cnt.

```python
# Now we'll build the Linear Model.
Y_train = Dt_train.pop('cnt')
X_train = Dt_train
```

```python
# Recursive the feature Elimination.
lm = LinearRegression()
lm.fit(X_train, Y_train)

rfe = RFE(lm, 15)
rfe = rfe.fit(X_train, Y_train)
```

```python
# Listing the variables selected in top 15 list.

list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Now Selecting the selected variables via RFE in COl list.

COL = X_train.columns[rfe.support_]
COL
```

```python
# Inspecting which columns has been rejected.

X_train.columns[~rfe.support_]
```

```python
# Generic Function for calculating the VIF of variables.
def cal_vif(bike_data):
    vif = pd.DataFrame()
    vif['Features'] = bike_data.columns
    vif['VIF'] = [variance_inflation_factor(bike_data.values, i) for i in range(bike_data.shape[1])]
    vif['VIF'] = round(vif['VIF'],2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif
```

```python
# Dataframe with RFE selected variables.

X_train_rfe = Dt_train[COL]
```

```python
# Calculating the VIF.
cal_vif(X_train_rfe)
```

##### From above, we can clearly see humidity displays high VIF value.

# 5: Building the Linear  Regression Model

```python
# Building thr 1st Linear Regression model.

X_train_lm_1 = sm.add_constant(X_train_rfe)
Lr_1 = sm.OLS(Y_train,X_train_lm_1).fit()
print(Lr_1.summary())
```

```python
# As we can see humidity shows high VIF values hence we can drop it.
X_train_new = X_train_rfe.drop(['humidity'], axis = 1)

# Running the function to calculate VIF for the new model.

cal_vif(X_train_new)
```

##### From above we can see that VIF values seems to be good now. But we'll see if we can reduced in further.

```python
X_train_lm_2 = sm.add_constant(X_train_new)
Lr_2 = sm.OLS(Y_train, X_train_lm_2).fit()
print(Lr_2.summary())
```

```python
# Now we can drop nov variable as it have high p-value.
X_train_new = X_train_new.drop(['nov'], axis = 1)

# Running the function to calculate VIF for the new model.

cal_vif(X_train_new)
```

##### After dropping few variables VIF does'nt seems to be affected much. Lets proceed further to observe better model.


```python
# Building 3rd Linear Regression Model.

X_train_lm_3 = sm.add_constant(X_train_new)
Lr_3 = sm.OLS(Y_train,X_train_lm_3).fit()
print(Lr_3.summary())
```

```python
# We can now drop dec variable as it have high p-value.

X_train_new = X_train_new.drop(['dec'], axis = 1)

# Running the function to calculate VIF for the new model.
cal_vif(X_train_new)
```

```python
# Building the 4th Linear Regression Model.
X_train_lm_4 = sm.add_constant(X_train_new)
Lr_4 = sm.OLS(Y_train,X_train_lm_4).fit()
print(Lr_4.summary())
```

```python
# We can drop jan variables now as it have high p-value.

X_train_new = X_train_new.drop(['jan'], axis = 1)

# Running the function to calculate VIF for the new model.
cal_vif(X_train_new)
```

```python
# Building the 5th Linear Regression Model.

X_train_lm_5 = sm.add_constant(X_train_new)
Lr_5 = sm.OLS(Y_train, X_train_lm_5).fit()
print(Lr_5.summary())
```

```python
# We'll drop july variables now as it have high p-value.

X_train_new = X_train_new.drop(['july'], axis = 1)

# Running the function to calculate VIF for the new model.

cal_vif(X_train_new)
```

```python
# Building 6th Linear Regression Model.
X_train_lm_6 = sm.add_constant(X_train_new)
Lr_6 = sm.OLS(Y_train,X_train_lm_6).fit()
print(Lr_6.summary())
```

##### We can consider the above model i.e Lr_6 as it have very low multicolinearity between the predictors and the p-values for all the predictors seems to be significant as well.

##### The F-statistics value i.e 248.4(which is greater than 1) and the p-value of 1.47e-186 which is almost equals to zero, indicates that the overall model is significant.

```python
# Checking the parameters and their coefficient values.
Lr_6.params
```

# 7: Making Predictions Using the Final Model

##### Now that we have fitted the model and checked the normality of error terms. It's time to move further and make predictions using the final model i.e Lr_6.

```python
# Applying Scaling on the test data_set.
var_num = ['temp', 'atemp', 'humidity', 'windspeed', 'cnt']
Dt_test[var_num] = scaler.transform(Dt_test[var_num])
Dt_test.head()
```

```python
Dt_test.describe()
```

```python
Y_test = Dt_test.pop('cnt')
X_test = Dt_test
```

```python
Col1 = X_train_new.columns

X_test = X_test[Col1]

# Adding constant variable to test dataframe

X_test_lm_6 = sm.add_constant(X_test)
```

```python
Y_pred = Lr_6.predict(X_test_lm_6)
```

```python
r2 = r2_score(Y_test, Y_pred)
round(r2, 4)
```

```python
# Calculating Adjusted-R^2 value for the test data_set.
Adjusted_r2 = round(1-(1-r2)*(X_test.shape[0]-1)/(X_test.shape[0]-X_test.shape[1]-1),4)
print(Adjusted_r2)
```
